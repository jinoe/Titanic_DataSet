```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 타이타닉 Train.csv 파일 데이터 전처리
train_df = pd.read_csv("train.csv")

#Name(이름), Ticket(티켓번호), Cabin 데이터는 생존률에 영향이 거의 없는것으로 간주하고 데이터 삭제
train_df = train_df.dropna(axis=1)
train_df = train_df.drop(['PassengerId','Name','Ticket','Fare'],axis=1)
from sklearn.preprocessing import LabelEncoder

# sklearn은 숫자형 데이터를 지원하기 때문에 object형의 데이터를 인코딩한다
# sklearn의 LabelEncoder 함수 이용 

encoder = LabelEncoder().fit(train_df.loc[:,'Sex'])
train_df.loc[:,'Sex'] = encoder.transform(train_df.loc[:,'Sex'])
#display(train_df)


#모델 학습
from sklearn.ensemble import RandomForestClassifier
display(train_df.drop(['Survived'], axis=1))
target = train_df[['Survived']].to_numpy() #target 설정
train_data = train_df.drop(['Survived'], axis=1).to_numpy() #target을 제외한 모든 열 선택
rf = RandomForestClassifier(n_estimators=1000,random_state=1).fit(train_data,target)
```

```python
test_data = pd.read_csv("test.csv")
test_data = test_data.dropna(axis=1)
test_data = test_data.drop(['PassengerId','Name','Ticket','Embarked'],axis=1)
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder().fit(test_data.loc[:,'Sex'])
test_data.loc[:,'Sex'] = encoder.transform(test_data.loc[:,'Sex'])
display(test_data)
test_data = test_data.to_numpy()

#print(rf.predict(test_data))
print(rf.score(train_data,target))
```

```python
test_target = pd.read_csv('gender_submission.csv')
test_target = test_target.iloc[:,1]
#display(test_target)

print("훈련데이터: ",rf.score(train_data,target))
print("테스트데이터: ",rf.score(test_data,test_target))
```

```python
test_df = pd.read_csv("test.csv")

print(test_df.iloc[:,0].shape)
print(rf.predict(test_data).shape)

result = {
    'PassengerId' : test_df.iloc[:,0],
    'Survived' : rf.predict(test_data)
}

result = pd.DataFrame(result)
display(result) 

result.to_csv('result.csv',index=False)
```
